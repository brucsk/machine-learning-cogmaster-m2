Similarly, $\boldsymbol{\sigma}_{b,\bullet} \in \{0,1\}^L$ denote the population activity configuration in time bin $b$, i.e. the $b$-th row of $A$.

#### Discretize the data

```python
# Create function to 
def bin_spike_trains(dt, X):
    """ Discretize spike times matrix.

    Inputs
    -------
        * dt : time bin (in seconds), float
        * X : spike times matrix, shape (n_spikes, 2) 
        X_{i,0} : i-th spike time (in seconds)
        X_{i,1} : neuron index corresponding to the i-th spike time

    Outputs
    -------
        * S : binarized spike-count matrix, shape (M,L)
        M : number of time bins
        L : number of neurons
    """
    # Compute total time
    T_total = round(np.max(X[:,0]))
    
    # Compute number of bins
    N_bins = int(T_total/dt)
    N_neurons = max(X[:,1]).astype(int) # the maximum index of the column #1 is the nb of neurons  
    
    # Initialize the spike-count matrices 
    S = np.zeros((N_bins, N_neurons), dtype=int)

    # Compute bin indices
    bin_idx = (X[:,0]/dt).astype(int)
    # Get neuron indices corresponding to the bin indices
    neuron_idx = (X[:,1]).astype(int) - 1
    
    np.add.at(S, (bin_idx, neuron_idx), 1) #not sure

    return S
        
# Define time bin width (delta)
delta = 0.05 # in seconds

S_pre = bin_spike_trains(delta, X_pre)
S_task = bin_spike_trains(delta, X_task)
S_post = bin_spike_trains(delta, X_post)
```

#### Binarize the data

```python
# Create function to 
def binarize_spike_counts(S):
    """ Binarize spike-count matrix to obtain a matrix with ones when there is at least one spike and 0 when there is none.
    
    Inputs
    -------
        * S : discretized spike data, shape (M, L)
        M : number of time bins
        L : number of neurons

    Outputs
    -------
        * Sig : sigma, binarized spike data, shape (M, L)
    """
    # Sig corresponds to the sign of S
    Sig = np.sign(S) 

    return Sig

Sig_pre = binarize_spike_counts(S_pre)
Sig_task = binarize_spike_counts(S_task)
Sig_post = binarize_spike_counts(S_post)
```

## 2 - Independent-site model

In this section, we introduce a simple **independent model** of neural population activity, where each neuron is represented as a binary random variable that can either fire ($\sigma_{b,i}=1$) or remain silent ($\sigma_{b,i}=0$) in each time bin, **independently** of all others
(*cf.* Section 5.2.1 in [1]).

This model neglects correlations between neurons but provides a useful **baseline**: it captures overall firing propensities and allows us to test whether population activity during **pre-task** and **post-task sleep** becomes more or less compatible with the statistics learned during the **task**.

Let $\boldsymbol{\sigma}_{b,\bullet} = (\sigma_{b,1}, \dots, \sigma_{b,L})$ denote the **population activity vector** at time bin $b$. The model assumes that for all $b$, the joint probability factorizes as

$$
P^{\text{ind}}(\boldsymbol{\sigma}_{b,\bullet})
= \prod_{i=1}^L P_i(\sigma_{b,i})
\quad \text{with} \quad
P_i(\sigma_{b,i}) = \frac{e^{h_i \sigma_{b,i}}}{1 + e^{h_i}}.
$$

The MLE estimates of the parameters $h_i$ can be obtained by the moment-matching condition:
$$
\langle \sigma_i \rangle_{\text{data}} = \frac{e^{h_i}}{1 + e^{h_i}}\ ,
$$
where $\langle \sigma_i \rangle_{\text{data}}$ are the empirical mean activities of each neuron $i$:
$$
\langle \sigma_i \rangle_{\text{data}} = \frac{1}{M} \sum_{b=1}^M \sigma_{b,i}
$$


#### ðŸŽ¯ Question 2a. Fitting the model

Starting from the binarized data matrices from (1.b), estimate the empirical mean activities $\langle \sigma_i \rangle_{\text{data}}$ for each neuron $i$ from the **task** session, and use the moment-matching condition to derive the corresponding fields $h_i$.

#### Derivation of local fields' theoretical expression 

We have the moment-matching condition given above:

$\langle \sigma_i \rangle_{\text{data}} = \frac{e^{h_i}}{1 + e^{h_i}} $

Let's isolate $h_i$:

$\langle \sigma_i \rangle_{\text{data}}(1 + e^{h_i}) = e^{h_i}$ 

$\Rightarrow \langle \sigma_i \rangle_{\text{data}} + \langle \sigma_i \rangle_{\text{data}} e^{h_i} = e^{h_i} $

$\Rightarrow \langle \sigma_i \rangle_{\text{data}} = e^{h_i}(1 - \langle \sigma_i \rangle_{\text{data}}) $ 

$ \Rightarrow e^{h_i} = \langle \sigma_i \rangle_{\text{data}}/(1 - \langle \sigma_i \rangle_{\text{data}}) $ 

$ \Rightarrow h_i = \ln(\langle \sigma_i \rangle_{\text{data}}/(1 - \langle \sigma_i \rangle_{\text{data}})) $ 

#### Using the empirical data to estimate the task local fields $h_i$

```python
def mean_activity(Sig):
    """
    Estimate empirical mean activity for each neuron i from a binarized data matrix.

    Inputs
    -------
        * Sig : sigma, binarized spike data, shape (M, L)
        M : number of time bins
        L : number of neurons

    Outputs
    -------
        * mean_val : empirical activities for each neuron averaged across time bins, shape (1, L)
    """
    # Compute mean across time bins for each neuron
    mean_val = Sig.mean(axis=0)
    # Get number of neurons
    L = Sig.mean(axis=0).shape[0]
    # Reshape from (,L) to (1,L) 
    mean_val = mean_val.reshape(1, L)
    
    return mean_val

def local_field(mean_val):
    """
    From the empirical mean activities for each neuron i of a given experimental session, compute corresponding local fields h_i 
    using the moment-matching condition.

    Inputs
    -------
        * mean_val : empirical mean activities for each neuron, shape (1, L)
        L : number of neurons

    Outputs
    -------
        * h : local fields for each neuron, shape (1, L)
    """
    h = np.log(mean_val/(1 - mean_val))
    return h

# Estimate mean activities for each neuron from the task session
mean_task = mean_activity(Sig_task)

# Get local field
h_task = local_field(mean_task)
```

```python
# Verify that the mean activity has the right shape
print(f"The shape of the mean activities from the task session is : {mean_task.shape}\nThe shape of the corresponding local fields is : {h_task.shape}")
```

#### ðŸŽ¯ Question 2b. Scoring configurations with the task model

Using the independent model fitted on the **task** data, the **log-probability** of each population activity configuration $\boldsymbol{\sigma}_{b,\bullet}$ reads:
$$
\log P^{\text{ind}}(\boldsymbol{\sigma}_{b,\bullet}) = \sum_{i=1}^{L} \left[ \sigma_{b,i} \log p_i + (1 - \sigma_{b,i})\log(1 - p_i) \right],
$$
where $p_i = \frac{e^{h_i}}{1 + e^{h_i}}$ are the firing probabilities from the task session.

**2.b.i** / Compute the log-probability for each time bin $b$ in the **pre-task**, **task**, and **post-task** sessions. Report the **mean and standard deviation** of log-probabilities for each session.

```python
# 2.b.i

def log_proba(Sig, h):
    """
    Compute the log probability of each population activity configuration using a previously fitted set of parameters. 

    Inputs
    -------
        * Sig : sigma, binarized spike data, shape (M, L)
        M : number of time bins
        L : number of neurons
        * h : fitted parameters for each neuron, shape (1, L)

    Outputs
    -------
        * log_proba : log-probability of population activity configuration for each time bin, shape (M, 1)
    """
    # Get number of time bins and neurons
    M, L = Sig.shape
    # Initialize variable containing firing probability for each neuron
    p = np.zeros((1,L))
    # Compute firing probabilities using the formula given above
    for i in range(L):
        p[0, i] = np.exp(h[0, i])/(1 + np.exp(h[0, i]))

    # Compute log probability 
    log_P = np.sum(Sig*np.log(p) + (1 - Sig)*np.log(1 - p), axis = 1)
    log_P = log_P.reshape((M, 1))

    return log_P

# Compute log probability for each session
log_proba_pre = log_proba(Sig_pre, h_task)
log_proba_task = log_proba(Sig_task, h_task)
log_proba_post = log_proba(Sig_post, h_task)
```

```python
print(log_proba_post.shape)
```

```python
print("Mean and standard deviation (std) of the log probability for each session:\n")

for session, log_P in zip(["Pre", "Task", "Post"], 
                      [log_proba_pre, log_proba_task, log_proba_post]):
    print(f"{session}: Mean = {np.mean(log_P):.3f} / Std = {np.std(log_P):.3f}")
```
