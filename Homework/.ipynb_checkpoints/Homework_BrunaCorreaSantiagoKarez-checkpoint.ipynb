#### Discretize the data

```python
# 1.a
def bin_spike_trains(dt, X):
    """ Discretize spike times matrix.

    Inputs
    -------
        * dt : time bin (in seconds), float
        * X : spike times matrix, shape (n_spikes, 2) 
        X_{i,0} : i-th spike time (in seconds)
        X_{i,1} : neuron index corresponding to the i-th spike time

    Outputs
    -------
        * S : spike-count matrix, shape (M,L)
        M : number of time bins
        L : number of neurons
    """
    # Compute total time
    T_total = round(np.max(X[:,0]))
    
    # Compute number of bins
    N_bins = int(T_total/dt)
    N_neurons = max(X[:,1]).astype(int) # the maximum index of the column #1 is the nb of neurons  
    
    # Initialize the spike-count matrices 
    S = np.zeros((N_bins, N_neurons), dtype=int)

    # Compute bin indices
    bin_idx = (X[:,0]/dt).astype(int)
    # Get neuron indices corresponding to the bin indices
    neuron_idx = (X[:,1]).astype(int) - 1
    
    np.add.at(S, (bin_idx, neuron_idx), 1) 

    return S
        
# Define time bin width (delta)
delta = 0.05 # in seconds

# Discretize spike times matrix for the three recording sessions
S_pre = bin_spike_trains(delta, X_pre)
S_task = bin_spike_trains(delta, X_task)
S_post = bin_spike_trains(delta, X_post)
```

#### Binarize the data

```python
# 1.b
def binarize_spike_counts(S):
    """ Binarize spike-count matrix to obtain a matrix with ones when there is at least one spike and 0 when there is none.
    
    Inputs
    -------
        * S : discretized spike data, shape (M, L)
        M : number of time bins
        L : number of neurons

    Outputs
    -------
        * Sig : sigma, binarized spike data, shape (M, L)
    """
    # Sig corresponds to the sign of S
    Sig = np.sign(S) 

    return Sig

# Binarize S matrices obtained above for the three recording sessions
Sig_pre = binarize_spike_counts(S_pre)
Sig_task = binarize_spike_counts(S_task)
Sig_post = binarize_spike_counts(S_post)
```

## 2 - Independent-site model

In this section, we introduce a simple **independent model** of neural population activity, where each neuron is represented as a binary random variable that can either fire ($\sigma_{b,i}=1$) or remain silent ($\sigma_{b,i}=0$) in each time bin, **independently** of all others
(*cf.* Section 5.2.1 in [1]).

This model neglects correlations between neurons but provides a useful **baseline**: it captures overall firing propensities and allows us to test whether population activity during **pre-task** and **post-task sleep** becomes more or less compatible with the statistics learned during the **task**.

Let $\boldsymbol{\sigma}_{b,\bullet} = (\sigma_{b,1}, \dots, \sigma_{b,L})$ denote the **population activity vector** at time bin $b$. The model assumes that for all $b$, the joint probability factorizes as

$$
P^{\text{ind}}(\boldsymbol{\sigma}_{b,\bullet})
= \prod_{i=1}^L P_i(\sigma_{b,i})
\quad \text{with} \quad
P_i(\sigma_{b,i}) = \frac{e^{h_i \sigma_{b,i}}}{1 + e^{h_i}}.
$$

The MLE estimates of the parameters $h_i$ can be obtained by the moment-matching condition:
$$
\langle \sigma_i \rangle_{\text{data}} = \frac{e^{h_i}}{1 + e^{h_i}}\ ,
$$
where $\langle \sigma_i \rangle_{\text{data}}$ are the empirical mean activities of each neuron $i$:
$$
\langle \sigma_i \rangle_{\text{data}} = \frac{1}{M} \sum_{b=1}^M \sigma_{b,i}
$$


#### üéØ Question 2a. Fitting the model

Starting from the binarized data matrices from (1.b), estimate the empirical mean activities $\langle \sigma_i \rangle_{\text{data}}$ for each neuron $i$ from the **task** session, and use the moment-matching condition to derive the corresponding fields $h_i$.

#### Derivation of local fields' theoretical expression 

We have the moment-matching condition given above:

$\langle \sigma_i \rangle_{\text{data}} = \frac{e^{h_i}}{1 + e^{h_i}} $

Let's isolate $h_i$:

$\langle \sigma_i \rangle_{\text{data}}(1 + e^{h_i}) = e^{h_i}$ 

$\Rightarrow \langle \sigma_i \rangle_{\text{data}} + \langle \sigma_i \rangle_{\text{data}} e^{h_i} = e^{h_i} $

$\Rightarrow \langle \sigma_i \rangle_{\text{data}} = e^{h_i}(1 - \langle \sigma_i \rangle_{\text{data}}) $ 

$ \Rightarrow e^{h_i} = \langle \sigma_i \rangle_{\text{data}}/(1 - \langle \sigma_i \rangle_{\text{data}}) $ 

$ \Rightarrow h_i = \ln(\langle \sigma_i \rangle_{\text{data}}/(1 - \langle \sigma_i \rangle_{\text{data}})) $ 

#### Using the empirical data to estimate the task local fields $h_i$

```python
def mean_activity(Sig):
    """
    Estimate empirical mean activity for each neuron i from a binarized data matrix.

    Inputs
    -------
        * Sig : sigma, binarized spike data, shape (M, L)
        M : number of time bins
        L : number of neurons

    Outputs
    -------
        * mean_val : empirical activities for each neuron averaged across time bins, shape (1, L)
    """
    # Compute mean across time bins for each neuron
    mean_val = Sig.mean(axis=0)
    # Get number of neurons
    L = Sig.mean(axis=0).shape[0]
    # Reshape from (,L) to (1,L) 
    mean_val = mean_val.reshape(1, L)
    
    return mean_val

def local_field(mean_val):
    """
    From the empirical mean activities for each neuron i of a given experimental session, compute corresponding local fields h_i 
    using the moment-matching condition.

    Inputs
    -------
        * mean_val : empirical mean activities for each neuron, shape (1, L)
        L : number of neurons

    Outputs
    -------
        * h : local fields for each neuron, shape (1, L)
    """
    # Correction: clamp to avoid infinities in logit
    eps = 1e-12
    mean_val_clip = np.clip(mean_val, eps, 1 - eps)
    # Logit 
    h = np.log(mean_val/(1 - mean_val_clip))
    return h

# Estimate mean activities for each neuron from the task session
mean_task = mean_activity(Sig_task)

# Get local field
h_task = local_field(mean_task)
```

```python
# Verify that the mean activity has the right shape
print(f"The shape of the mean activities from the task session is : {mean_task.shape}\nThe shape of the corresponding local fields is : {h_task.shape}")
```

#### üéØ Question 2b. Scoring configurations with the task model

Using the independent model fitted on the **task** data, the **log-probability** of each population activity configuration $\boldsymbol{\sigma}_{b,\bullet}$ reads:
$$
\log P^{\text{ind}}(\boldsymbol{\sigma}_{b,\bullet}) = \sum_{i=1}^{L} \left[ \sigma_{b,i} \log p_i + (1 - \sigma_{b,i})\log(1 - p_i) \right],
$$
where $p_i = \frac{e^{h_i}}{1 + e^{h_i}}$ are the firing probabilities from the task session.

**2.b.i** / Compute the log-probability for each time bin $b$ in the **pre-task**, **task**, and **post-task** sessions. Report the **mean and standard deviation** of log-probabilities for each session.

```python
# 2.b.i

def log_proba(Sig, h):
    """
    Compute the log probability of each population activity configuration using a previously fitted set of parameters. 

    Inputs
    -------
        * Sig : sigma, binarized spike data, shape (M, L)
        M : number of time bins
        L : number of neurons
        * h : fitted parameters for each neuron, shape (1, L)

    Outputs
    -------
        * log_proba : log-probability of population activity configuration for each time bin, shape (M, 1)
    """
    # Get number of time bins and neurons
    M, L = Sig.shape
    # Initialize variable containing firing probability for each neuron
    p = np.zeros((1,L))
    # Compute firing probabilities using the formula given above
    for i in range(L):
        p[0, i] = np.exp(h[0, i])/(1 + np.exp(h[0, i]))

    # Compute log probability 
    log_P = np.sum(Sig*np.log(p) + (1 - Sig)*np.log(1 - p), axis = 1)
    log_P = log_P.reshape((M, 1))

    return log_P

# Compute log probability for each session
log_proba_pre = log_proba(Sig_pre, h_task)
log_proba_task = log_proba(Sig_task, h_task)
log_proba_post = log_proba(Sig_post, h_task)
```

```python
print("Mean and standard deviation (std) of the log probability for each session:\n")

for session, log_P in zip(["Pre", "Task", "Post"], 
                      [log_proba_pre, log_proba_task, log_proba_post]):
    print(f"{session}: Mean = {np.mean(log_P):.3f} / Std = {np.std(log_P):.3f}")
```

#### 2.b.iii - Interpretation 

In the Bayesian inference framework, the log-likelihood score informs us about the probability that the observed data points are generated by a given model. Here, the model is an independent Bernouilli process, which assumes that neurons fire independently of one another.

The parameters fitted from the independent-site model provide log-probability distributions that are very similar across the three recording sessions (pre-task, task and post-task), with a mean of approximately -13.5 for the pre-task session and -13.7 for both the post-task and task sessions. The standard deviation is approximately 3.2, 3.4 and 3.3 for pre-task, task and post-task sessions respectively, indicating a broad overlap of the three distributions. This means that if we consider only the single-neuron level, the population activity patterns seem to be stable across sessions.

Surprisingly, we observe that the mean log-probability for the pre-task population activity pattern is superior to that of the task, even though the model parameters $h_i$ were fitted using the empirical mean activities of the task session. However, the effect size (Cohen's d) is negligible (d $\approx$ 0.06), suggesting that this effect might only stem from artifacts such as noise. From these observations, we can infer that using a model that considers the variables (here, neurons) as independent, and therefore the population-level distribution as the product of single-neuron distributions, does not capture the similarity of post- and task with respect to pre- and task. This model is underfitted, and we might seek to refine it to capture more information about the population activity, for instance, by considering pair-wise interactions between neurons which cause them to not fire independently. 

Furthermore, the strong assumption that neurons fire independently from one another is biologically implausible, as neurons are coupled by synapses: the electrical activity of a pre-synaptic neuron affects the electrical activity of a post-synaptic neuron. The recordings coming from a multi-electrode array, it is probable that several recorded neurons are synaptically coupled.

#### Compute cosine similarity between local field vectors

```python
# 2.c.ii

def cos_similarity(h1, h2):
    """
    Compute the cosine similarity between a pair of local field vectors (h1, h2) using a function from the library Scikit-learn.

    Inputs
    -------
        * h1 : fitted parameters for each neuron on one session (ex: pre-task, task or post-task), shape (1, L)
        * h2 : fitted parameters for each neuron on another session, shape (1, L)

    Outputs
    -------
        * cos_sim : cosine similarity between the two vectors h1 and h2, defined as h_1.h_2/(||h1||.||h2||), float
    """
    # Compute cosine similarity (we get the element [0][0] to have a float insted of 1-element array)
    cos_sim = cosine_similarity(h1, h2)[0][0]
    
    return cos_sim

# Calcule cosine similarity for pairs of local field vectors
cos_sim_pre_task = cos_similarity(h_pre, h_task)
cos_sim_post_task = cos_similarity(h_post, h_task)
cos_sim_pre_post = cos_similarity(h_pre, h_post)

# Print results
for pair, cos_sim in zip(["(h_pre, h_task)", "(h_post, h_task)", "(h_pre, h_post)"], 
                      [cos_sim_pre_task, cos_sim_post_task, cos_sim_pre_post]):
    print(f"cosine similarity{pair} = {cos_sim:.3f}")
```

## 3 - Gaussian model with pairwise correlations

The independent model captures individual firing propensities but ignores **pairwise correlations** between neurons. We now extend our analysis to a **Gaussian model** that accounts for these correlations, providing a richer description of population activity.  (*cf.* Section 5.1 in [1])

For z-scored neural activity $Y_{b,\bullet} = (Y_{b,1}, \dots, Y_{b,L})$ with zero mean and unit variance, the Gaussian model assumes:
$$
P^{\text{Gauss}}(Y_{b,\bullet}) = \frac{1}{(2\pi)^{L/2} |C|^{1/2}} \exp\left(-\frac{1}{2} Y_{b,\bullet}^T \, T \, Y_{b,\bullet}\right),
$$

where $T = C^{-1}$ is the **precision matrix** (the inverse of the covariance matrix $C$).

Because the data are z-scored (zero mean, unit variance), the model is fully characterized by the **off-diagonal elements** $T_{ij}$ ($i \neq j$), which encode **effective pairwise couplings** between neurons. The **maximum likelihood estimate (MLE)** of the precision matrix is obtained by inverting the empirical covariance matrix: $T = C^{-1}$. A large positive (negative) coupling $T_{ij}$ indicates that neurons $i$ and $j$ tend to be active together (in opposite patterns) more often than expected by chance.

#### üéØ Question 3a. Data preprocessing and correlation structure

Starting from the **spike-count matrices** $S_{\text{pre}}$, $S_{\text{task}}$, and $S_{\text{post}}$ (not the binarized versions!), **z-score** each matrix to obtain standardized activity matrices $Y$ with zero mean and unit variance per neuron, and compute the **sample correlation matrices** for each session:
$$
C = \frac{1}{M} Y^T Y
$$

*Note: See Tutorial 3, Q1b-c for reference on z-scoring the spike-count matrices (not the binarized ones!) and correlation computation.*

```python
# 3.a
def compute_z_score(S):
    """ Computes z-scored matrix Y given a spike-count matrix S 
    
    Inputs
    -------
        * S : binarized spike-count matrix, shape (M, L)
        M : number of time bins
        L : number of neurons

    Outputs
    -------
        * Y : z-scored data where each column has zero mean and unit variance, shape (M, L)
    """
    M = len(S)
    p = np.sum(S, axis = 0)/M
    var = 1/(M-1)*np.sum((S - p)**2, axis = 0)
    Y = (S - p)/np.sqrt(var)
    return Y

def compute_corr_matrix(Y):
    """ Computes correlation matrix out of z-scored data Y

    Inputs
    -------
        * Y : z-scored data, shape (M, L)
        M : number of time bins
        L : number of neurons

    Outputs
    -------
        * C : correlation matrix, shape (L, L)
    """
    M = len(Y)
    C = (1/M)*(Y.T @ Y)
    return C

# Z-score spike-count matrices
Y_pre = compute_z_score(S_pre)
Y_task = compute_z_score(S_task)
Y_post = compute_z_score(S_post)

# Get correlation matrices 
C_pre = compute_corr_matrix(Y_pre)
C_task = compute_corr_matrix(Y_task)
C_post = compute_corr_matrix(Y_post)
```


#### üéØ Question 3b. Scoring configurations with the Gaussian model

**3.b.i** / Compute the **precision matrix** $T_{\text{task}} = C_{\text{task}}^{-1}$ from the task session.

```python
# 3.b.i
def precision_matrix(C):
    """ Computes precision matrix T from correlation matrix.

    Inputs
    -------
        * C : correlation matrix, shape (L, L)

    Outputs
    -------
        * T : maximum likelihood estimate (MLE) of the precision matrix, shape (L, L)
    """
    T = inv(C)
    return T

T_task = precision_matrix(C_task)

print(f"T_task:{T_task}")
```

```python
# Sanity check: see if T is symetric 
def isSymmetric(mat, N, thres):
    for i in range(N):
        for j in range(N):
            if (mat[i][j] - mat[j][i]>= thres):
                return False
    return True

# Define arbitrary precision threshold (we probably won't have T_{i,j} exactly equal to T_{j,i} bc numerical imprecision)
thres = 10**-14
# Get order of T
N_t = T_task.shape[0]

# Use function to test if T is symetric
if isSymmetric(T_task, N_t, thres):
    print("The matrix T_task is symetric")
else:
    print("The matrix T_task is not symetric")
```

**3.b.ii** / For each time bin $b$ in all three sessions, compute the log-probability under the Gaussian model fitted on the task:
  $$
  \log P^{\text{Gauss}}(Y_{b,\bullet}) = -\frac{1}{2} \sum_{i \neq j} T_{ij} Y_{b,i} Y_{b,j}
  $$
  **Note:** We exclude diagonal terms $T_{ii}$ since the z-scoring already accounts for individual firing rates. Furthermore, the normalization constant is omitted as it is the same for all configurations.

Report the **mean and standard deviation** of log-probability scores for each session.

```python
# 3.b.ii

def log_prob_gauss(Y, T):
    """ Computes precision matrix T from correlation matrix.

    Inputs
    -------
        * Y : z-scored data, shape (M, L)
            M : number of time bins
            L : number of neurons
        * T : maximum likelihood estimate (MLE) of the precision matrix, shape (L, L)

    Outputs
    -------
        * logP_gauss : log-probability under Gaussian model for each time bin, shape (1, M)
        * mean_LPG : mean of logP_gauss, float
        * std_LPG : standard deviation of logP_gauss, float
    """
    # We exclude diagonal terms T_{i,i} 
    T_no_diag = T.copy()
    np.fill_diagonal(T_no_diag, 0)
    # Initialize the vector to store log probabilities 
    M, L = Y.shape
    logP_gauss = np.zeros((M,1))
    # Compute log probability for each population z-scored activity at time bin b
    for b in range(M):
        logP_gauss[b, 0] = - (1/2)*(Y[b,:] @ T_no_diag @ Y[b,:])
    # Compute mean and standard deviation
    mean_LPG = np.mean(logP_gauss)
    std_LPG = np.std(logP_gauss)
    
    return logP_gauss, mean_LPG, std_LPG

# Compute log-probability under the Gaussian model fitted with on the task for each session
logP_gauss_pre, mean_LPG_pre, std_LPG_pre = log_prob_gauss(Y_pre, T_task)
logP_gauss_task, mean_LPG_task, std_LPG_task = log_prob_gauss(Y_task, T_task)
logP_gauss_post, mean_LPG_post, std_LPG_post = log_prob_gauss(Y_post, T_task)

print("Mean and standard deviation (std) of the log probability under Gaussian model for each session:\n")

for session, log_P in zip(["Pre", "Task", "Post"], 
                      [logP_gauss_pre, logP_gauss_task, logP_gauss_post]):
    print(f"{session}: Mean = {np.mean(log_P):.3f} / Std = {np.std(log_P):.3f}")
```

This is more coherent than the findings obtained with the independent model, in which the mean log-probabilities were very similar across the three recording sessions. We can deduce that the Gaussian model is richer than the independent-firing model, as it is able to capture the differences in population activity patterns between task and non-task sessions, and therefore better capture the structure of the data. 

A hypothesis to explain these observations is that what differentiates the task session from the other sessions arises from subpopulation-level coordination, which are not apparent when considering only single-neuron firing activity.

**3.c.iv** / Interpretation: Do the pairwise coupling patterns during post-task sleep resemble those during the task more than pre-task sleep does? What does this tell you about the consolidation of interaction structure between neurons during sleep?

#### 3.c.iv - Interpretation

We observe a stronger cosine similarity between $J_{post}$ and $J_{task}$ ($\approx$ 0.6) than between $J_{pre}$ and $J_{task}$ ($\approx$ 0.4). This means that the interactions between neurons during the task more closely resemble those after the task than before, pointing towards the idea that some  task-related neural interactions are kept or reproduced when the rat sleeps after the task. 

We can interpret the similarity of the pairwise couplings between task and post-task population activity as the result of cell assembly formation and replay: subpopulations of neurons are coactivated during the task, and the same groups are reactivated during sleep to consolidate the memory of the task (Peyrache *et al.*, 2010, *J. Comput Neurosci*). The fact that memories are encoded by changing the strength of connections between pairs aligns with the seminal Hebb's postulate (*"neurons that fire together wire together"*; Hebb, 1949). Nowadays, we have access to an increasing amount of computational tools allowing to effectively measure these coactivations in animals. Thanks to the Gaussian modelling approach, we are able to identify assembly replay in the prefrontal cortex during sleep. 

Furthermore, we also have a strong cosine similarity between $J_{pre}$ and $J_{post}$, which is likely due to similarities in patterns during sleep. To deepen our interpretation, we should take our analysis further and extract which of the elements $J_{task}^{i,j}$ are more similar to $J_{post}^{i,j}$ than to $J_{pre}^{i,j}$ to know which of the neurons are part of the cell assembly whose synaptic strength changes during memory consolidation. 

Another adjacent reflection is that the reinforced pattern that we observe is limited to the neurons that we record. We have a finite amount of data, therefore we might not be capturing the assemblies might include other neurons.
