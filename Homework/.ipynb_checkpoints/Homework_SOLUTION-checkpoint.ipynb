## 1 - From spikes to binary activity

#### ðŸŽ¯ Question 1a. Discretization

- For each condition (`pre`, `task`, and `post`), and using time bins of width $\Delta t = 0.05\,\mathrm{s}$ (50 ms), construct the spike-count matrices $ S \in \mathbb{N}^{M \times L}, $
where each **row** corresponds to a time bin and each **column** to a neuron. Each entry $S_{b,i}$ represents the **number of spikes** emitted by neuron $i$ within time bin $b$.

```python
def discretize_data(data, dt=0.1, verbose=False):
    """
    Discretize spike events into an MÃ—L spike-count matrix.

    Parameters
    ----------
    data : array-like, shape (n_spikes, 2)
        Columns: [time_sec, neuron_id]. Time is in seconds, already shifted to start at 0.
    dt : float
        Bin width in seconds (default: 0.1 s).
    verbose : bool
        If True, print additional information.
    Returns
    -------
    S : np.ndarray, shape (M, L)
        Spike-count matrix. Rows = time bins, columns = neurons (1..L mapped to 0..L-1).
    """
    times = data[:, 0] # First column: spike times (in seconds)
    ids   = data[:, 1].astype(int)  # Second column: neuron IDs (assumed to be 1..L)

    t_max = times.max()
    M = int(np.floor(t_max / dt)) + 1       # number of time bins
    L = int(ids.max())                      # assume IDs are 1..L

    S = np.zeros((M, L), dtype=int)

    # Bin index for each spike
    b = np.floor(times / dt).astype(int)
    # Column index for each spike (0-based)
    c = ids - 1

    # Accumulate counts
    np.add.at(S, (b, c), 1)

    if verbose:
        print(f"Data discretized into {M} time bins and {L} neurons.")
        print(f"Spike-count matrix shape: {S.shape} (M Ã— L)")
    return S

dt = 0.05 # Bin width in seconds (50 ms)
S_pre  = discretize_data(X_pre,  dt=dt, verbose=True)
S_task = discretize_data(X_task, dt=dt, verbose=True)
S_post = discretize_data(X_post, dt=dt, verbose=True)
```

#### ðŸŽ¯ Question 1b. Binarisation
Then,  for each condition, **binarize** the data to obtain
$$
\sigma_{b,i} =
\begin{cases}
1 & \text{if}\ S_{b,i} >0 \\
0 & \text{otherwise.}
\end{cases}\ .
$$
Note that the binarized activity matrix $\boldsymbol{\sigma} \in \{0,1\}^{M \times L}$ has the same dimensions as $S$.

We will use the symbol $\boldsymbol{\sigma}_{\bullet,i} \in \{0,1\}^M$ to denote the activity of neuron $i$ across all $M$ time bins, i.e. the $i$-th column of $\sigma$. Similarly, $\boldsymbol{\sigma}_{b,\bullet} \in \{0,1\}^L$ denote the population activity configuration in time bin $b$, i.e. the $b$-th row of $\sigma$.


```python
def binarize_spike_counts(S):
    """
    Binarize spike-count matrix: all values > 0 become 1.
    """
    S_bin = (S > 0).astype(int)
    S_bin[S_bin > 1] = 1
    return S_bin

S_bin_pre = binarize_spike_counts(S_pre)
S_bin_task = binarize_spike_counts(S_task)
S_bin_post = binarize_spike_counts(S_post)
```

## 2 - Independent-site model

In this section, we introduce a simple **independent model** of neural population activity, where each neuron is represented as a binary random variable that can either fire ($\sigma_{b,i}=1$) or remain silent ($\sigma_{b,i}=0$) in each time bin, **independently** of all others
(*cf.* Section 5.2.1 in [1]).

This model neglects correlations between neurons but provides a useful **baseline**: it captures overall firing propensities and allows us to test whether population activity during **pre-task** and **post-task sleep** becomes more or less compatible with the statistics learned during the **task**.

Let $\boldsymbol{\sigma}_{b,\bullet} = (\sigma_{b,1}, \dots, \sigma_{b,L})$ denote the **population activity vector** at time bin $b$. The model assumes that for all $b$, the joint probability factorizes as

$$
P^{\text{ind}}(\boldsymbol{\sigma}_{b,\bullet})
= \prod_{i=1}^L P_i(\sigma_{b,i})
\quad \text{with} \quad
P_i(\sigma_{b,i}) = \frac{e^{h_i \sigma_{b,i}}}{1 + e^{h_i}}.
$$

The MLE estimates of the parameters $h_i$ can be obtained by the moment-matching condition:
$$
\langle \sigma_i \rangle_{\text{data}} = \frac{e^{h_i}}{1 + e^{h_i}}\ ,
$$
where $\langle \sigma_i \rangle_{\text{data}}$ are the empirical mean activities of each neuron $i$:
$$
\langle \sigma_i \rangle_{\text{data}} = \frac{1}{M} \sum_{b=1}^M \sigma_{b,i}
$$


#### ðŸŽ¯ Question 2a. Fitting the model

Starting grom the binarized data matrices from (1.b), estimate the empirical mean activities $\langle \sigma_i \rangle_{\text{data}}$ for each neuron $i$ from the **task** session, and use the moment-matching condition to derive the corresponding fields $h_i$.

```python
def bernoulli_params_from_binned(S_bin):
    """
    Compute per-neuron Bernoulli probabilities and logistic fields from a binned binary spike matrix.

    Parameters
    ----------
    S_bin : np.ndarray, shape (M, L)
        Binary spike-count matrix with M time bins (rows) and L neurons (columns).
        Entries should be 0/1 or boolean; each column corresponds to a neuron activity vector.

    Returns
    -------
    p : np.ndarray, shape (L,)
        Empirical firing probability for each neuron:
        p_i = (# of bins where neuron i is active) / M. Values in [0, 1].
        These correspond to <sigma_i>_data

    h : np.ndarray, shape (L,)
        Independent-model fields (logits) computed as:
        h_i = log(p_i / (1 - p_i)).
        To avoid infinities when p_i is exactly 0 or 1, probabilities are clipped to a small
        epsilon before taking the logit.

    Notes
    -----
    - The returned `p` can be used directly to score binary configurations under the
      independent Bernoulli model.
    - Clipping uses eps = 1e-12 to ensure numerical stability.
    """
    M, L = S_bin.shape
    # Bernoulli per-bin probability p_i
    p = np.count_nonzero(S_bin, axis=0) / M

    # Clamp to avoid infinities in logit
    eps = 1e-12
    p_clip = np.clip(p, eps, 1 - eps)
    h = np.log(p_clip / (1.0 - p_clip))  # logit

    return p, h

p_task, h_task = bernoulli_params_from_binned(S_bin_task)
```

#### ðŸŽ¯ Question 2b. Scoring configurations with the task model

Using the independent model fitted on the **task** data, the **log-probability** of each population activity configuration $\boldsymbol{\sigma}_{b,\bullet}$ reads:
$$
\log P^{\text{ind}}(\boldsymbol{\sigma}_{b,\bullet}) = \sum_{i=1}^{L} \left[ \sigma_{b,i} \log p_i + (1 - \sigma_{b,i})\log(1 - p_i) \right],
$$
where $p_i = \frac{e^{h_i}}{1 + e^{h_i}}$ are the firing probabilities from the task session.

**2.b.i** / Compute the log-probability for each time bin $b$ in the **pre-task**, **task**, and **post-task** sessions. Report the **mean and standard deviation** of log-probabilities for each session.

```python
def score_session(S_bin, p):
    """
    Compute log-probability scores for all configurations in a session.

    Parameters
    ----------
    S_bin : np.ndarray, shape (M, L)
        Binary spike matrix (M = bins, L = neurons).
    p : np.ndarray, shape (L,)
        Bernoulli probabilities from the task model.

    Returns
    -------
    scores : np.ndarray, shape (M,)
        Log-probabilities of each configuration (per time bin).
    """
    # Precompute log-probabilities for active and silent states
    logP1 = np.log(p)
    logP0 = np.log(1 - p)
    # Compute log-probability for each time bin using vectorized operations
    return (S_bin @ logP1 + (1 - S_bin) @ logP0)

scores_pre  = score_session(S_bin_pre, p_task)
scores_task = score_session(S_bin_task, p_task)
scores_post = score_session(S_bin_post, p_task)
```

**2.b.ii** / Plot and compare the **distributions** of log-probability scores (one for each time bin) across the three sessions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = 1/2.54 

def plot_score_distributions(scores_pre, scores_task, scores_post, nbins=50, score_range=None):

    plt.figure(figsize=(10 * cm, 8 * cm))
    bins = np.linspace(min(scores_pre.min(), scores_task.min(), scores_post.min()),
                       max(scores_pre.max(), scores_task.max(), scores_post.max()), nbins)
    
    if score_range is not None:
        bins = np.linspace(score_range[0], score_range[1], nbins)
        
    plt.hist(scores_pre,  bins=bins, histtype='step', lw=.5, color='C0', label='Pre-task sleep', density=True)
    plt.hist(scores_task, bins=bins, histtype='step', lw=1, color='C1', label='Task', density=True)
    plt.hist(scores_post, bins=bins, histtype='step', lw=.5, color='C2', label='Post-task sleep', density=True)

    plt.title('Log-probability score distributions', fontsize=12)
    plt.xlabel('Log-probability score', fontsize=10)
    plt.ylabel('Density', fontsize=10)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.legend(frameon=False)
    sns.despine()

    plt.show()

plot_score_distributions(scores_pre, scores_task, scores_post)


print("\nâœ… Configurations scored under independent Bernoulli model from task data (log-probabilities).")
print(f"  Pre-task sleep : {scores_pre.mean():.3f} Â± {scores_pre.std():.3f}")
print(f"  Task           : {scores_task.mean():.3f} Â± {scores_task.std():.3f}")
print(f"  Post-task sleep: {scores_post.mean():.3f} Â± {scores_post.std():.3f}")
```

**2.c.ii** /   Calculate and report the **cosine similarity** for the following pairs of local field vectors: $(\mathbf{h}_{\text{pre}}, \mathbf{h}_{\text{task}})$, $(\mathbf{h}_{\text{post}}, \mathbf{h}_{\text{task}})$,  $(\mathbf{h}_{\text{post}}, \mathbf{h}_{\text{pre}})$.

â„¹ï¸ **Reminder**: The cosine similarity measures the alignment between two vectors, ranging from $-1$ (opposite directions) to $+1$ (same direction), and is defined as:
  $$
  \text{cosine similarity}(v_1, v_2) = \frac{v_1 \cdot v_2}{\|v_1\| \, \|v_2\|}
  $$
  where $v_1 \cdot v_2 $ is the dot product and $\|v\| $ is the Euclidean norm.
  

```python
def cosine_similarity(v1, v2, eps=1e-12):
    """
    Compute cosine similarity between two vectors.
    
    Measures the cosine of the angle between vectors, ranging from -1 to 1:
      +1: vectors point in same direction (perfectly similar)
       0: vectors are orthogonal (uncorrelated)
      -1: vectors point in opposite directions
    
    Parameters
    ----------
    v1, v2 : array-like, shape (N,)
        Input vectors to compare (must have same length)
    eps : float, optional (default=1e-12)
        Small value to avoid division by zero
    
    Returns
    -------
    float
        Cosine similarity in [-1, 1]
    
    Notes
    -----
    Formula: cos_sim = (v1 Â· v2) / (||v1|| * ||v2||)
    """
    # Convert to numpy arrays
    v1 = np.asarray(v1)
    v2 = np.asarray(v2)
    
    # Compute: (v1 Â· v2) / (||v1|| * ||v2||)
    numerator = np.dot(v1, v2)
    denominator = np.linalg.norm(v1) * np.linalg.norm(v2) + eps
    
    return numerator / denominator

print(f"Cosine similarity (h_pre vs h_task):  {cosine_similarity(h_pre, h_task):.4f}")
print(f"Cosine similarity (h_post vs h_task): {cosine_similarity(h_post, h_task):.4f}")
print(f"Cosine similarity (h_post vs h_pre):  {cosine_similarity(h_post, h_pre):.4f}")
```

## 3 - Gaussian model with pairwise correlations

The independent model captures individual firing propensities but ignores **pairwise correlations** between neurons. We now extend our analysis to a **Gaussian model** that accounts for these correlations, providing a richer description of population activity.  (*cf.* Section 5.1 in [1])

For z-scored neural activity $Y_{b,\bullet} = (Y_{b,1}, \dots, Y_{b,L})$ with zero mean and unit variance, the Gaussian model assumes:
$$
P^{\text{Gauss}}(Y_{b,\bullet}) = \frac{1}{(2\pi)^{L/2} |C|^{1/2}} \exp\left(-\frac{1}{2} Y_{b,\bullet}^T \, T \, Y_{b,\bullet}\right),
$$

where $T = C^{-1}$ is the **precision matrix** (the inverse of the covariance matrix $C$).

Because the data are z-scored (zero mean, unit variance), the model is fully characterized by the **off-diagonal elements** $T_{ij}$ ($i \neq j$), which encode **effective pairwise couplings** between neurons. The **maximum likelihood estimate (MLE)** of the precision matrix is obtained by inverting the empirical covariance matrix: $T = C^{-1}$. A large positive (negative) coupling $T_{ij}$ indicates that neurons $i$ and $j$ tend to be active together (in opposite patterns) more often than expected by chance.

#### ðŸŽ¯ Question 3a. Data preprocessing and correlation structure

Starting from the **spike-count matrices** $S_{\text{pre}}$, $S_{\text{task}}$, and $S_{\text{post}}$ (not the binarized versions!), **z-score** each matrix to obtain standardized activity matrices $Y$ with zero mean and unit variance per neuron, and compute the **sample correlation matrices** for each session:
$$
C = \frac{1}{M} Y^T Y
$$

*Note: See Tutorial 3, Q1b-c for reference on z-scoring the spike-count matrices (not the binarized ones!) and correlation computation.*

```python
def zscore_data(S):
    """
    Z-score each column (neuron) of the spike-count matrix S.

    Parameters
    ----------
    S : np.ndarray, shape (M, L)
        Spike-count matrix. Rows = time bins, columns = neurons.

    Returns
    -------
    Y : np.ndarray, shape (M, L)
        Z-scored spike-count matrix.
    """
    means = S.mean(axis=0)
    stds  = S.std(axis=0, ddof=1) # ddof=1 for a normalisation by M-1
    # Avoid division by zero for neurons with zero variance
    stds_safe = np.where(stds > 0, stds, 1.0)

    Y = (S - means) / stds_safe
    return Y

Y_pre  = zscore_data(S_pre)
Y_task = zscore_data(S_task)
Y_post = zscore_data(S_post)

def sample_correlation(Y):
    """
    Compute sample correlation matrix C = (1/M) Y^T Y from z-scored data.
    Y : (M, L) with zero-mean, unit-variance columns.
    Returns C : (L, L), symmetric PSD.
    """
    M = Y.shape[0]
    return (Y.T @ Y) / M

C_pre = sample_correlation(Y_pre)
C_task = sample_correlation(Y_task)
C_post = sample_correlation(Y_post)
```


#### ðŸŽ¯ Question 3b. Scoring configurations with the Gaussian model

**3.b.i** / Compute the **precision matrix** $T_{\text{task}} = C_{\text{task}}^{-1}$ from the task session.

```python
T_task = np.linalg.inv(C_task)
```

**3.b.ii** / For each time bin $b$ in all three sessions, compute the log-probability under the Gaussian model fitted on the task:
  $$
  \log P^{\text{Gauss}}(Y_{b,\bullet}) = -\frac{1}{2} \sum_{i \neq j} T_{ij} Y_{b,i} Y_{b,j}
  $$
  **Note:** We exclude diagonal terms $T_{ii}$ since the z-scoring already accounts for individual firing rates. Furthermore, the normalization constant is omitted as it is the same for all configurations.

Report the **mean and standard deviation** of log-probability scores for each session.

```python
def gaussian_logpdf_precision(Z, T, include_const=False):
    """
    Compute log-probability of z-scored configurations under a Gaussian model
    using the precision matrix formulation.

    The Gaussian log-probability (unnormalized) is:
        log P(z) = -0.5 * z^T * T * z + const

    where T = C^{-1} is the precision matrix (inverse covariance).

    Since z-scoring already accounts for individual firing rates (diagonal terms),
    we exclude T_ii from the quadratic form and only compute pairwise interactions.

    Parameters
    ----------
    Z : np.ndarray, shape (M, L)
        Z-scored activity matrix

    T : np.ndarray, shape (L, L)
        Precision matrix (inverse of covariance matrix) from the task session.

    include_const : bool, optional (default=False)
        If True, include the normalization constant:
        -0.5 * [log|C| + L*log(2Ï€)]
        where |C| is the determinant of the covariance matrix.

    Returns
    -------
    logp : np.ndarray, shape (M,)
        Log-probability score for each time bin (configuration).
        Higher values indicate configurations more consistent with the task model.

    Notes
    -----
    We use np.einsum for computationally efficient computation of the quadratic form. Learn to use this!
    See: https://numpy.org/doc/stable/reference/generated/numpy.einsum.html

    The einsum operation 'mi,ij,mj->m' computes:
        result[m] = sum_i sum_j Z[m,i] * T_offdiag[i,j] * Z[m,j]
    which is equivalent to the diagonal of (Z @ T_offdiag @ Z.T).
    """
    T_offdiag = T.copy()
    np.fill_diagonal(T_offdiag, 0)

    # This computes for each row (time bin) m the quadratic form: Z[m,:] @ T_offdiag @ Z[m,:]
    # Resulting shape: (M,)
    quad = np.einsum('mi,ij,mj->m', Z, T_offdiag, Z)
    
    logp = -0.5 * quad

    # Optionally add the normalization constant
    if include_const:
        # Compute log|C| = -log|T| using sign-safe log-determinant
        sign_T, logdet_T = np.linalg.slogdet(T)
        logdet_C = -logdet_T

        L = Z.shape[1]

        logp -= 0.5 * (logdet_C + L * np.log(2.0 * np.pi))

    return logp


# Score all sessions using the task precision matrix
# This evaluates how well task-learned couplings explain activity in each state
lp_task = gaussian_logpdf_precision(Y_task, T_task, include_const=False)
lp_pre  = gaussian_logpdf_precision(Y_pre,  T_task, include_const=False)
lp_post = gaussian_logpdf_precision(Y_post, T_task, include_const=False)

# Report summary statistics
print("\nMean log-probability scores for each session (under the task precision model):")
print(f"  Pre-task sleep : {lp_pre.mean():.3f} Â± {lp_pre.std():.3f}")
print(f"  Task           : {lp_task.mean():.3f} Â± {lp_task.std():.3f}")
print(f"  Post-task sleep: {lp_post.mean():.3f} Â± {lp_post.std():.3f}")
```

**3.b.iv** / Intepretation: How do these results compare with those from the independent model (Question 2b)? Does accounting for correlations change which session's activity is best captured by the task model?

#### ðŸŽ¯ Question 3c. Comparing coupling matrices across behavioral states

The precision matrix $T$ (or equivalently, its negative $J = -T$, called the **coupling matrix**) encodes the interaction structure between neurons. We now ask: how does this interaction structure differ across behavioral states?

**3.c.i** / Compute the coupling matrices for all three sessions:
  $$
  J_{\text{pre}} = -C_{\text{pre}}^{-1}, \quad J_{\text{task}} = -C_{\text{task}}^{-1}, \quad J_{\text{post}} = -C_{\text{post}}^{-1}
  $$

```python
J_task = - T_task 
J_pre = -np.linalg.inv(C_pre)
J_post = -np.linalg.inv(C_post)
```

**3.c.ii** / Visualize the three coupling matrices as heatmaps (with diagonal elements set to zero for clarity).

```python
def plot_couplings(C, title="Correlation Matrix"):
    """
    Plot the correlation matrix C as a heatmap, with diagonal set to zero.
    """
    C_plot = C.copy()
    np.fill_diagonal(C_plot, 0)
    plt.figure(figsize=(8 * cm, 6 * cm))
    sns.heatmap(C_plot, cmap='coolwarm', center=0, square=True, vmin=-.1, vmax=.1, cbar_kws={"shrink": 0.8})
    plt.title(title, fontsize=12)
    plt.xlabel("Neuron index", fontsize=10)
    plt.ylabel("Neuron index", fontsize=10)
    plt.xticks(fontsize=8)
    plt.yticks(fontsize=8)
    plt.tight_layout()
    plt.show()

# plot heatmaps of the precision matrices
plot_couplings(J_pre, title="Precision Matrix (Pre-task sleep)")
plot_couplings(J_task, title="Precision Matrix (Task session)")
plot_couplings(J_post, title="Precision Matrix (Post-task sleep)")
```

**3.c.iii** / Compute the **cosine similarity** (see definition above) between the linearised off-diagonal elements of each pair of coupling matrices: ($J_{\text{pre}}$ and $J_{\text{task}}$), ($J_{\text{post}}$ and $J_{\text{task}}$), and ($J_{\text{pre}}$ and $J_{\text{post}}$).

```python
def offdiag_vec(J):
    """Vectorize the strictly upper-triangular (off-diagonal) entries of J."""
    iu = np.triu_indices_from(J, k=1)
    return J[iu]

def pairwise_scores(JA, JB, eps=1e-12):
    """
    Compare two coupling matrices using several metrics on their off-diagonals.
    Returns a dict with Pearson r, cosine sim, and (normalized) Frobenius distance.
    """
    a = offdiag_vec(JA)
    b = offdiag_vec(JB)

    # Cosine similarity (magnitude-sensitive version of correlation)
    cos = float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b) + eps))

    return cos

scores_pre_task  = pairwise_scores(J_pre,  J_task)
scores_post_task = pairwise_scores(J_post, J_task)
scores_pre_post  = pairwise_scores(J_pre,  J_post)

print("Similarity(J_pre,  J_task):", scores_pre_task)
print("Similarity(J_post, J_task):", scores_post_task)
print("Similarity(J_pre,  J_post):", scores_pre_post)
```

**3.c.iv** / Interpretation: Do the pairwise coupling patterns during post-task sleep resemble those during the task more than pre-task sleep does? What does this tell you about the consolidation of interaction structure between neurons during sleep?

 ðŸ‘‰ **Answer** /  Yes, post-task sleep coupling patterns resemble task patterns more closely than pre-task sleep does. Post-task sleep is more similar to the task (0.557 vs 0.355), indicating that the pairwise interaction structure shifts toward task-learned patterns after the task.

The two sleep sessions are most similar to each other (0.606), suggesting they share a common baseline. However, post-task sleep shows a clear shift toward the task coupling structure, while remaining intermediate between baseline sleep and task patterns. This provides evidence for consolidation of neuronal interaction patterns during sleep. The task modifies how neurons coordinate their activity (pairwise couplings), and these learned interaction patterns are partially retained and reactivated during post-task sleep -- consistent with the two-stage model of memory consolidation.


## Further comments

#### 1. On the bin size choice
The choice of bin size is a **methodological decision** that must balance statistical, biological, and practical considerations. There is no single "correct" bin size -- different choices emphasize different aspects of the data. Smaller bins provide finer temporal resolution but lead to sparser data (many bins with zero spikes), while larger bins improve statistics but may blur rapid dynamics and merge distinct activity patterns. A bin size of 50 ms represents a compromise between two constraints:
- **Lower bound (temporal resolution)**: Bins should be larger than the synaptic integration timescale (~10-20 ms) to capture coordinated population activity emerging from multi-step cortical processing ([Markram et al., Science, 1997](https://www.science.org/doi/10.1126/science.275.5297.213)).
- **Upper bound (binarization error, if binarizing)**: Bins should be small enough that multiple spikes per bin are rare. With PFC firing rates of ~1-5 Hz, a 50 ms bin corresponds to ~0.05-0.25 expected spikes per bin, making the probability of multiple spikes low. This ensures that binarization (mapping all spike counts > 0 to 1) introduces minimal information loss.

Replay sequences in rodent PFC last 40-200 ms total, with precise spike timing on ~10-20 ms scales (Peyrache et al., 2009). Our 50 ms bins are intermediate between single-neuron integration timescales and full replay event durations. This binning is appropriate for comparing population-level statistics (firing rates, correlations) across behavioral states, though too coarse to resolve fine temporal order within individual replay events. Standard bin sizes in population coding studies range from 10-100 ms.

The plot below compares raw spike counts $S_{b,i}$ vs. binarized activity $\sigma_{b,i}$, confirming that multiple spikes per bin are indeed rare.

```python
import matplotlib.pyplot as plt
cm = 1/2.54 # centimeters in inches

def plot_binarization(S, S_bin, title="Binarization"):
    """
    Plot distributions before and after binarization.
    """

    plt.figure(figsize=(8 * cm, 5 * cm))
    plt.hist(S.flatten(), bins=np.arange(S.max() + 2) - 0.5, alpha=0.5, label="$S_{b,i}$", edgecolor='w', density=True)
    plt.hist(S_bin.flatten(), bins=np.arange(3) - 0.5, alpha=0.5, label=r"$\sigma_{b,i}$", edgecolor='w', density=True)
    plt.xlabel("Spike count / activity per bin", fontsize=10)
    plt.ylabel("Frequency", fontsize=10)
    plt.legend(fontsize=10)
    plt.tight_layout()
    ax = plt.gca()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.show()

    print(f"\nDistribution before binarization ({title}):")
    unique, counts = np.unique(S, return_counts=True)
    for u, c in zip(unique, counts):
        print(f"  {u} spikes: {c} ({c/S.size:.1e})")

    print("\nâœ… Data binarized (all spike counts > 0 set to 1).")
    unique, counts = np.unique(S_bin, return_counts=True)
    print(f"Distribution after binarization ({title}):")
    for u, c in zip(unique, counts):
        print(f" Value {u}: {c} ({c/S_bin.size:.4f})")

plot_binarization(S_pre, S_bin_pre, title="Pre-task sleep")
plot_binarization(S_task, S_bin_task, title="Task (maze)")
plot_binarization(S_post, S_bin_post, title="Post-task sleep")
```

#### 2. Statistical vs Practical Significance (Independent Model)

Although the distributions appear to overlap substantially (as seen in the histogram), this does **not** mean they are statistically identical. With thousands of time bins per session (pre: 14,179, task: 11,308, post: 13,171), even small differences in means can be highly statistically significant.

**Hypothesis Testing Framework:**
- **Null hypothesis ($H_0$)**: The distributions are identical (samples come from the same population)
- **Alternative hypothesis ($H_1$)**: The distributions differ significantly
- **p-value**: Probability of observing such extreme differences if $H_0$ were true
- **Significance level (Î±)**: Typically 0.05 or 0.001; if $p < \alpha$, we reject $H_0$

The overlapping distributions tell us about the **effect size** (small, $\sim$ 0.2 log-probability difference), while statistical tests reveal whether these differences are **reliable** (unlikely due to chance). In neuroscience, such small but consistent differences can be biologically meaningful, especially when accumulated over many observations.

To verify statistical significance, we can perform:
- [Kolmogorov-Smirnov test](https://en.wikipedia.org/wiki/Kolmogorov%E2%80%93Smirnov_test): Tests if entire distributions differ (shape, location, scale)
- [Mann-Whitney U test](https://en.wikipedia.org/wiki/Mann%E2%80%93Whitney_U_test): Non-parametric test for different medians (robust to outliers)

The key insight: The independent Bernoulli model reveals subtle but potentially significant differences between behavioral states. However, the small effect size (difference < 2% of the mean) suggests that individual firing rates alone provide limited discrimination power, motivating the need for models that capture pairwise correlations (Section 3).

```python
from scipy import stats
import numpy as np

def compare_distributions(s1, s2, label1, label2):
    """
    Compare two distributions using statistical tests.
    Hâ‚€: The distributions are identical
    """
    # Statistical tests
    # Kolmogorovâ€“Smirnov test
    ks_stat, ks_pval = stats.ks_2samp(s1, s2)
    # Mannâ€“Whitney U test
    mw_stat, mw_pval = stats.mannwhitneyu(s1, s2, alternative='two-sided')
    
    # Effect size (Cohen's d)
    pooled_std = np.sqrt((s1.std()**2 + s2.std()**2) / 2)
    cohens_d = (s1.mean() - s2.mean()) / pooled_std
    
    print(f"\n{label1} vs {label2}:")
    print(f"  Mean diff: {s1.mean() - s2.mean():.4f}")
    print(f"  KS test: p = {ks_pval:.2e} â†’ {'REJECT Hâ‚€ âœ“' if ks_pval < 0.001 else 'Cannot reject Hâ‚€'}")
    print(f"  MW test: p = {mw_pval:.2e} â†’ {'REJECT Hâ‚€ âœ“' if mw_pval < 0.001 else 'Cannot reject Hâ‚€'}")

# Run comparisons
print("=" * 50)
print("STATISTICAL TESTS (Hâ‚€: distributions are identical)")
print("=" * 50)

compare_distributions(scores_pre, scores_task, "Pre-task", "Task")
compare_distributions(scores_post, scores_task, "Post-task", "Task")
compare_distributions(scores_pre, scores_post, "Pre-task", "Post-task")
```
